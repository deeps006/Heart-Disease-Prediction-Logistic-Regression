```python
import pandas as pd
heart = pd.read_csv(r"E:\Data science\heart.csv")
heart
```

```python
x=heart[["thalach","age","thal","trestbps"]]
x
```

```python
y=heart["target"]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred)
```

```python
#input values = Maximum heart rate,age,Thalassemia test result,Resting Blood Pressure
import numpy as np
def heart_disease(patient_data):
    patient_array = np.array(patient_data)
    prediction = model.predict(patient_array)
    probability = model.predict_proba(patient_array)
    for i,p in enumerate(prediction):
        print(f"Prediction:{"Heart Disease" if p==1 else "No Heart Disease"}")
        print(f"Probability (No/Yes):{probability[i]}")
mhr=int(input("Enter the  Maximum heart rate:"))
age=int(input("Enter the age:"))
tt=int(input("Enter the Thalassemia test result(1/2/3...etc):"))
rbp=int(input("Enter your Resting time BP:"))
values=[[mhr,age,tt,rbp]]
heart_disease(values)
```
